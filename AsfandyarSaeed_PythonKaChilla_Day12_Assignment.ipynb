```python
import numpy as np
import pandas as pd
from numpy import *
import matplotlib.pyplot as pt
import seaborn as sns
```

```python
land = pd.read_csv("land_used.csv")
```

```python
#land.goupby (["Item",])

sns.barplot(x="Item",y="Value",data=land,hue = "Item", order =["Other land","Agricultural land"],color="red",estimator=mean,palette='pastel',
            ci=None, saturation =0.8)
```

```python
# now grouping values 
land.groupby (["Item","Year"]).mean()
```
